```python
def fix_reversed_counts(row, max_counter):
    counter = row['ENTRIES'] - row['PREV_ENTRIES']
    if counter < 0:
        counter = -counter
    if counter > max_counter:
        print(row['ENTRIES'], row['PREV_ENTRIES'])
        return 0
    return counter

_=turnstile_df.apply(fix_reversed_counts, axis=1, max_counter = 1000000)
```
